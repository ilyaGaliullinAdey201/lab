2.2.1 Напишите программирование на Python для отображения гистограммы популярности языков программирования.  
Пример данных:
Языки программирования: Java, Python, PHP, JavaScript, C #, C ++
Популярность: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
langs = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C ++']
students = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
ax.bar(langs, students)
ax.minorticks_on()
ax.grid(which = 'major', color = 'r', linewidth = 1)
ax.grid(which = 'minor', color = 'k', linestyle = ':')
ax.set_xlabel('Languages')
ax.set_ylabel('Popularity')
ax.set_title('Popularuty of Programming Language \n Worldwide, Oct 2017 compared to a year ago')
plt.show()
```

2.3.1. Напишите Python для программирования, чтобы создать круговую диаграмму популярности языков программирования.  
Пример данных:
Языки программирования: Java, Python, PHP, JavaScript, C #, C ++
Популярность: 22.2, 17.6, 8.8, 8, 7.7, 6.7


```
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('equal')
explode = [0.1,0,0,0,0,0]
langs = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
students = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
ax.pie(students, labels = langs, autopct = '%1.2f%%', explode=explode,shadow=True, wedgeprops={'lw':1,'edgecolor':"k"}, rotatelabels=True)
ax.axis("equal")
plt.show()
```

2.4.1. Напишите программу на Python, чтобы нарисовать график рассеяния со случайным распределением по осям X и Y, нанесенный на график друг против друга.  

```
import numpy as np
import numpy.random as rand
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

x = np.random.uniform(-2, 4, 180)
y = np.random.uniform(-2, 4, 180)
plt.xlim(-3, 3)
plt.ylim(-3, 3)
fig = plt.scatter(x, y, c = 'r', s = 10)
plt.xlabel('X')
plt.ylabel('Y')
plt.tick_params(direction = 'in', top = True, right = True)
plt.show()
```

```
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.pyplot import figure

df = pd.read_csv('Mall_Customers.csv', sep = ',', parse_dates=True)
fig = plt.figure(figsize=(15,5))
ax = fig.add_axes([0, 0, 1, 1])
mf = df[df['Genre'] == 'Male']
mf = mf.sort_values('Age')
mf = mf.groupby(['Age']).mean()
x1 = mf.index
y1 = mf[['Annual Income (k$)']]
ax.set_xticks([0,5,10,15,20,25,30,35,40,45,50,55,60,65,70])
l1 = ax.plot(x1, y1, '--')
plt.style.use("ggplot")
ff = df[df['Genre'] == 'Female']
ff = ff.sort_values('Age')
ff = ff.groupby(['Age']).mean()
x2 = ff.index
y2 = ff[['Annual Income (k$)']]
l2= ax.plot(x2, y2, '--')
ax.set_title("Графики годового дохода по возрасту")
ax.set_xlabel('Возраст')
ax.set_ylabel('Размер годового дохода')
ax.legend(labels = ('Males', 'Females'), loc = 'lower right')
```

4.	Реализовать цветовую сетку данных

```
from google.colab import files
uploaded = files.upload()
for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(name = fn, length = len(uploaded[fn])))
```

```
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.pyplot import figure

df = pd.read_csv('Mall_Customers.csv', sep = ',', parse_dates=True)
plt.figure(figsize=(10,5))
mf = df[df['Genre'] == 'Male']
mf = mf.sort_values('Age')
y1 = mf[['Spending Score (1-100)']]
ff = df[df['Genre'] == 'Female']
ff = ff.sort_values('Age')
y2 = ff[['Spending Score (1-100)']]
fig, ax1 = plt.subplots()
ax1.pcolormesh(y1)
fig, ax2 = plt.subplots()
ax2.pcolormesh(y2, cmap='hot')
ax1.set_title("Рейтинг трат мужчин")
ax2.set_title("Рейтинг трат женщин")
ax1.set_ylabel('Рейтинг трат')
ax2.set_ylabel('Рейтинг трат')
fig.set_facecolor('floralwhite')
ax1.set_facecolor('seashell')
plt.show()
```

Анализируя цветовую сетку данных по рейтингу трат мужчин и женщин, можно сделать вывод, что самый часто встречаюшийся уровень от 59 до 65, а у женщин от 29 до 29.

5.	Построить 3D график.

```
from google.colab import files
uploaded = files.upload()
for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(name = fn, length = len(uploaded[fn])))
```

```
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.pyplot import figure

df = pd.read_csv('Mall_Customers.csv', sep = ',', parse_dates=True)
fig = plt.figure(figsize=(10,5))
ax = fig.add_axes([0, 0, 1, 1])
mf = df[df['Genre'] == 'Male']
mf = mf.sort_values('Age')
x1 = mf[['Age']]
y1 = mf[['Annual Income (k$)']]
z1 = mf[['Spending Score (1-100)']]
ax.set_xticks([0,5,10,15,20,25,30,35,40,45,50,55,60,65,70])
ax = fig.add_subplot(111, projection='3d')
l1 = ax.scatter(x1, y1, z1, c = 'fuchsia', s = 40)
ax.set_facecolor('wheat')
ff = df[df['Genre'] == 'Female']
ff = ff.sort_values('Age')
x2 = ff[['Age']]
y2 = ff[['Annual Income (k$)']]
z2 = ff[['Spending Score (1-100)']]
l2= ax.scatter(x2, y2, z2, c = 'cyan', s = 40)
ax.set_title("3-d диаграмма рассеивания годового дохода \n и рейтинга тратпо возрасту")
ax.set_xlabel('Возраст')
ax.set_ylabel('Размер годового дохода')
ax.set_zlabel('Рейтинг трат')
ax.legend(labels = ('Males', 'Females'), loc = 'lower right')
```

На тайкой 3-д диаграмме хорошо видно, какие параметры встречаются чаще всего. Например, пересечение возраста от 25 до 35, годовой доход от 100 до 120 и рейтинг трат 60-80. Людей обоих полов с такими параметрами больше всех.
